# Nuclear Power Plant Electrical Energy Prediction

## Problem Statement:

* Predict the Electrical Energy generated by a Nuclear Power Plant.
### Some of the attributes of the dataset:
1. Ambient Temperature

2. Exhaust Vaccume

3. Ambient Pressure

4. Relative Humidity

5. Electrical Energy Output


## Solution:

* As the dependent variable is of numeric value type, we can use Regression Algorithms

* Some of the availble classification algorithms are: 
    1. Linear Regression

    2. Multiple Linear Regression

    3. Polynomial Regression

    4. Decision Tree Regressor

    5. Random Forest Regressor

    6. SVM Regressor



* **In our Solution we are using Random Forest Regresser**

## Importing libraries

```python
#data handling libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#machine learning libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.ensemble import RandomForestRegressor
```

## Importing dataset

```python
dataset = pd.read_csv('Nuclear Power Plant Production.csv')
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
X[0]
```

```python
y[0]
```

## Splitting the dataset into Training & Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=21)
```

## Random Forest Regression Training on Training set

```python
regressor = RandomForestRegressor(n_estimators=20)
```

```python
regressor.fit(X_train,y_train)
```

## Predicting

```python
y_pred = regressor.predict(X_test)
```

## Visualizing Test Set results

```python
plt.plot(y_pred, label='Predicted Data',c='g')
plt.plot(y_test, label='Train Data',c='b')
plt.plot(abs((y_test) -(y_pred)), label='Error',c='r')
plt.legend()
plt.show()
```

# Model Evaluation

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
